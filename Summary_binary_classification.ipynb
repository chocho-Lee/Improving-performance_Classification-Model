### Result
#### 모든 검사항목 사용, 모든 검사항목에 대한 클래스 가중치 적용, 의학적 중요도 높은 항목만 사용, 의학적 중요도+클래스 가중치 적용한 4가지 모델을 "기준모델"로, 위의 4가지의 기준모델 각각에 대한 튜닝 결과 모델을 "비교모델"로 표현한다.<br>
#### negative보다는 positive에 대한 예측률이 높아야 하는 모델이라는 점을 감안하여 각 모델을 분석하면 비교모델에 비해 기준모델들에서 TP가 더 높게 출력되므로 일차적으로 기준모델이 조금 더 유효한 모델이라고 해석할 수 있다.<br>
#### 기준모델과 비교모델을 통틀어 모든 검사항목을 사용하여 클래스 가중치를 적용한 모델에서 TP가 가장 높은 것으로 확인된다.<br>
#### GridSearchCV를 활용하여 F1 score를 기준으로 한 최적의 성능을 탐색했으나, 하이퍼파라미터를 튜닝한 모델에서 더 좋은 성능이 나올 것이라는 예상과 다른 결과가 나온 점에 대해 GridSearhCV의 파라미터를 재조정하거나 RandomizedSearchCV 방식을 적용하여 추적할 필요성이 있다.<br>
#### 참고 :: null값이 많은 항목의 경우 각 열의 평균으로 채웠으며 기계학습 모델은 다른 모델들에 비해 좋은 성능을 보이는 것으로 경험한 XGBoost를 선택하였다.
